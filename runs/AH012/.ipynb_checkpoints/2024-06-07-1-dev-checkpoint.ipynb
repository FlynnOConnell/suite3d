```python
@njit(parallel = True, nogil = True, cache = True)
def compute_pixel_snr(f, frame_rate):
    df = n.diff(f)
    df_over_f = df / f.mean(axis=)
    abs_d_dff = n.abs(n.diff(df_over_f,axis=1))
    noise_level = n.nanmedian(abs_d_dff, axis=1)
    noise_level = noise_level / frame_rate
    return noise_level
```
